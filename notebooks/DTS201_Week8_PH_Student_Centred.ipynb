## 1. What Is Business Intelligence (BI)?

**Business Intelligence (BI)** refers to tools, techniques, and processes that help organisations:

- **Collect** data from different sources.  
- **Transform** raw data into meaningful information.  
- **Visualise** trends, patterns, and KPIs.  
- **Support decisions** such as pricing, resource allocation, marketing, or risk management.

Examples of BI questions:

- â€œWhich product categories generate the highest revenue?â€  
- â€œWhich region is underperforming compared to others?â€  
- â€œWhat is the trend of monthly sales over the last year?â€  

In this lab, you will work with a **simple sales or performance dataset** and simulate a mini BI dashboard.

### 3.1 Parsing Dates and Basic Cleaning

```
# If your dataset has a 'Date' column, it may be loaded as text (string)
# We convert it to a proper datetime type for time-based analysis
if 'Date' in df.columns:
    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # errors='coerce' sets invalid dates to NaT (missing)

    # Create additional useful time-based columns if needed
    df['Year'] = df['Date'].dt.year          # Extract the year from the date
    df['Month'] = df['Date'].dt.to_period('M')  # Extract year-month as a period (e.g. 2024-01)

# Display updated data types to confirm that 'Date' was converted successfully
df.dtypes  # Check which columns are datetime, numeric, or object (text)
```

## 4. Identifying KPIs (Key Performance Indicators)

In BI dashboards, we often focus on a small number of **KPIs** to summarise performance.  

Common examples in a sales context:

- **Total Revenue** â€“ overall sales amount.  
- **Total Quantity Sold** â€“ total units sold.  
- **Average Order Value** â€“ average sales per transaction.  
- **Profit Margin** â€“ profit as a percentage of revenue (if profit and cost are available).

We will compute some basic KPIs from the dataset.


```
# Choose column names for key numeric measures
# Replace 'Sales_Amount' and 'Quantity' with actual columns from your dataset
sales_col = "Sales_Amount"  # TODO: change to your sales or revenue column
qty_col = "Quantity"        # TODO: change to your quantity column (if available)

# Compute total revenue
total_revenue = df[sales_col].sum()  # Sum of all sales amounts

# Compute total quantity if the column exists
total_quantity = df[qty_col].sum() if qty_col in df.columns else None  # Sum of quantities if available

# Compute number of transactions (rows)
num_transactions = len(df)  # Count of rows/records

# Compute average order value (if both sales and transactions are valid)
avg_order_value = total_revenue / num_transactions  # Average sales amount per transaction

# Print KPI values
print("=== Basic KPIs ===")
print(f"Total Revenue: {total_revenue:,.2f}")
if total_quantity is not None:
    print(f"Total Quantity Sold: {total_quantity:,.0f}")
print(f"Number of Transactions: {num_transactions}")
print(f"Average Order Value: {avg_order_value:,.2f}")
```

## 5. Aggregating KPIs by Category (Region, Product, etc.)

Dashboards typically show KPIs **broken down by categories**, such as region, branch, or product type.

We will compute **total revenue** (and optionally quantity) by a chosen category.


```
# Choose a categorical column for grouping
# Example options: 'Region', 'Branch', 'Product_Category'
group_col = "Region"  # TODO: change to a real categorical column from your dataset

# Group the data by the selected category and compute total revenue and quantity
group_summary = df.groupby(group_col).agg(
    Total_Revenue=(sales_col, "sum"),   # Sum of sales for each category
    Average_Revenue=(sales_col, "mean") # Average sales per transaction for each category
)

# If quantity column exists, also compute total quantity
if qty_col in df.columns:
    group_summary["Total_Quantity"] = df.groupby(group_col)[qty_col].sum()

# Sort groups by total revenue in descending order
group_summary = group_summary.sort_values(by="Total_Revenue", ascending=False)

# Display the summary table
group_summary
```

## 6. Visualizing KPIs with Bar Charts

Bar charts are useful for comparing KPIs **across categories** (e.g. revenue by region).

```
# Reset the index so that 'group_col' becomes a normal column for plotting
group_summary_reset = group_summary.reset_index()  # Converts index to a column

# Plot a bar chart of total revenue by category
plt.figure(figsize=(8, 5))  # Set the figure size

# Use seaborn barplot to show Total_Revenue by group_col
sns.barplot(data=group_summary_reset, x=group_col, y="Total_Revenue")

# Add a title and axis labels
plt.title(f"Total Revenue by {group_col}")
plt.xlabel(group_col)
plt.ylabel("Total Revenue")

# Rotate x-axis labels if they are long
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

## 7. Visualizing Trends Over Time (Line Chart)

If your dataset contains a `Date` column, you can track KPIs **over time**, such as monthly revenue.

We will compute **total revenue per month** and plot it as a line chart.


```
# Check if 'Month' column exists (created earlier from 'Date')
if 'Month' in df.columns:
    # Group by Month and aggregate sales
    monthly_summary = df.groupby('Month').agg(
        Total_Revenue=(sales_col, 'sum')  # Total revenue per month
    ).reset_index()

    # Sort by Month to ensure the line chart follows time order
    monthly_summary = monthly_summary.sort_values(by="Month")

    # Plot the monthly revenue trend
    plt.figure(figsize=(9, 5))  # Set figure size

    # Use seaborn lineplot to draw the trend of total revenue over time
    sns.lineplot(data=monthly_summary, x='Month', y='Total_Revenue', marker="o")

    # Add title and labels
    plt.title("Monthly Total Revenue Over Time")
    plt.xlabel("Month")
    plt.ylabel("Total Revenue")

    # Rotate x-axis labels to make them readable
    plt.xticks(rotation=45)

    # Adjust layout and display plot
    plt.tight_layout()
    plt.show()
else:
    print("No 'Month' column found. Ensure 'Date' was properly parsed and 'Month' was created.")
```

## 8. Adding Simple Filters (Simulating Dashboard Slices)

BI dashboards typically allow users to **filter** data by region, product, or time period.

We can simulate this behaviour using **pandas queries** and re-running our summaries and plots.


```
# Choose a specific category value to filter on
# Example: filter to only one region, such as 'North' or 'Lagos'
filter_value = "North"  # TODO: change to a real category value from your dataset

# Filter the dataset for the chosen category
df_filtered = df[df[group_col] == filter_value]  # Keep only rows where group_col equals filter_value

# Compute total revenue for this filtered subset
filtered_revenue = df_filtered[sales_col].sum()

print(f"Total revenue for {group_col} = {filter_value}: {filtered_revenue:,.2f}")

# If 'Month' is available, also compute monthly revenue for this subset
if 'Month' in df_filtered.columns:
    monthly_filtered = df_filtered.groupby('Month').agg(
        Total_Revenue=(sales_col, 'sum')
    ).reset_index().sort_values(by='Month')

    # Plot the monthly trend for the filtered category
    plt.figure(figsize=(9, 5))
    sns.lineplot(data=monthly_filtered, x='Month', y='Total_Revenue', marker="o")
    plt.title(f"Monthly Revenue for {group_col} = {filter_value}")
    plt.xlabel("Month")
    plt.ylabel("Total Revenue")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
