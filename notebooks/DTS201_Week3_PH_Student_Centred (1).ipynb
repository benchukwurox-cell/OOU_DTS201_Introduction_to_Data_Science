## 1. Getting Started: What Is EDA?

**Exploratory Data Analysis (EDA)** is the process of **looking inside the data** before doing any complex modelling.

In EDA, we usually:

- Ask: *â€œWhat does each row represent?â€*  
- Check: *â€œAre there missing values or strange values?â€*  
- Summarise: *â€œWhat are the typical values (mean, median, range)?â€*  
- Visualise: *â€œWhat patterns, trends, or groups do we see?â€*  

In this notebook, you will practise EDA on a real dataset of your choice.

## 5. Exploring Distributions of Numerical Variables

We start by looking at **one numerical column** at a time.

Common questions:

- What is a typical value (mean/median)?  
- Is the distribution skewed?  
- Are there obvious outliers?


```
# TODO: Replace with a real numeric column from your dataset
num_col = "NUMERIC_COLUMN"  # e.g. "G3", "age", "score"

# Drop missing values in the chosen column
num_data = df[num_col].dropna()

# Print basic statistics
print(num_col)
print("Count:", num_data.count())
print("Mean:", num_data.mean())
print("Median:", num_data.median())
print("Standard deviation:", num_data.std())
print("Minimum:", num_data.min())
print("Maximum:", num_data.max())
```

```
# Plot a histogram with a density curve for the numeric column
plt.figure(figsize=(7, 4))
sns.histplot(num_data, kde=True)
plt.title(f"Distribution of {num_col}")
plt.xlabel(num_col)
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 6. Exploring Categorical Variables

For categorical variables, we are interested in **counts** and **proportions** for each category.

Examples: gender, region, class, product category, etc.


```
# TODO: Replace with a real categorical column
cat_col = "CATEGORICAL_COLUMN"  # e.g. "sex", "school", "region"

# Show value counts (how many times each category appears)
df[cat_col].value_counts()
```

```
# Plot a bar chart of category counts
plt.figure(figsize=(7, 4))
sns.countplot(x=df[cat_col])
plt.title(f"Counts of {cat_col}")
plt.xlabel(cat_col)
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. Exploring Relationships Between Two Numerical Variables

Sometimes we want to know how two numerical variables move together.

Example questions:

- *â€œDo students who study more hours tend to have higher scores?â€*  
- *â€œIs income related to spending score?â€*


```
# TODO: Replace with two real numeric columns
x_col = "NUMERIC_X"  # e.g. "studytime"
y_col = "NUMERIC_Y"  # e.g. "G3"

# Keep only rows where both are not missing
pair_df = df[[x_col, y_col]].dropna()

# Scatter plot with regression line
plt.figure(figsize=(7, 4))
sns.scatterplot(data=pair_df, x=x_col, y=y_col)
sns.regplot(data=pair_df, x=x_col, y=y_col, scatter=False, color="red")
plt.title(f"{y_col} vs {x_col}")
plt.tight_layout()
plt.show()

# Compute correlation (if enough data)
if len(pair_df) > 1:
    corr_val = pair_df[x_col].corr(pair_df[y_col])
    print(f"Correlation between {x_col} and {y_col}: {corr_val:.3f}")
```

## 8. Comparing a Numerical Variable Across Categories

We can also compare the distribution of a numerical variable across different categories.

Examples:

- Exam score by gender.  
- Income by region.  
- Spending score by customer segment.


```
# TODO: Replace with a numeric column and a categorical column
num_col2 = "NUMERIC_COLUMN"      # e.g. "G3"
group_col = "GROUP_COLUMN"       # e.g. "sex"

plt.figure(figsize=(7, 4))
sns.boxplot(data=df, x=group_col, y=num_col2)
plt.title(f"{num_col2} by {group_col}")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
