## 5. Cleaning and Standardizing Data

### 5.1 Handling Missing Values

```python
# Example: Fill missing numerical values with median
df_csv['Age'] = df_csv['Age'].fillna(df_csv['Age'].median())
```

### 5.2 Correcting Wrong Data Types

```python
# Convert 'rating' column to numeric
df_csv['rating'] = pd.to_numeric(df_csv['rating'], errors='coerce')

# Clean and convert 'rating_count' column to numeric
df_csv['rating_count'] = df_csv['rating_count'].str.replace(',', '', regex=False).astype(float)
```

```python
print(df_csv[['rating', 'rating_count']].dtypes)
```

```python
import random

gender_options = ['Male', 'Female', 'Other']
df_csv['Gender'] = [random.choice(gender_options) for _ in range(len(df_csv))]

# Display the DataFrame head with the new 'Gender' column
df_csv.head()
```

### 5.3 Standardizing Category Labels

```python
# Example: Standardize variations of 'Male'
df_csv['Gender'] = df_csv['Gender'].str.lower().replace({'m': 'male', 'male ': 'male'})
```

### 5.4 Removing Duplicates

```python
# Count duplicates
df_csv.duplicated().sum()

# Remove them
df_csv = df_csv.drop_duplicates()
```
