## 4. Generating Summary Statistics

We start EDA by computing summary statistics for **numerical** and **categorical** features.

Some common functions are:

- `df.describe()` â€“ summary of numerical columns.  
- `df['column'].mean()` â€“ mean of a column.  
- `df['column'].median()` â€“ median of a column.  
- `df['column'].value_counts()` â€“ frequency table for a categorical column.


```
# Generate descriptive statistics for all numerical columns
df.describe()  # Shows count, mean, std, min, max, and quartiles for numerical features
```

```
# Example: compute specific statistics for one numerical column
# Replace 'Score' with the name of a numerical column from your dataset
num_col = "Score"  # TODO: change this to a real numerical column name

# Calculate and print the mean of the selected column
print("Mean of", num_col, ":", df[num_col].mean())  # Average value

# Calculate and print the median of the selected column
print("Median of", num_col, ":", df[num_col].median())  # Middle value when sorted

# Calculate and print the standard deviation of the selected column
print("Standard Deviation of", num_col, ":", df[num_col].std())  # Measure of spread
```

## 5. Visualizing Categorical Data with Bar Charts

Bar charts are useful for showing the frequency of categories (e.g., gender, department, grade).

```
# Create a bar chart for the selected categorical column
plt.figure(figsize=(8, 4))  # Set the size of the plot (width, height)

# Use seaborn's countplot to automatically count each category in the column
sns.countplot(x=df[cat_col])  # x-axis shows categories; y-axis shows their counts

# Add a title and labels to make the plot easier to interpret
plt.title(f"Bar Chart of {cat_col}")  # Title of the plot
plt.xlabel(cat_col)  # Label for the x-axis
plt.ylabel("Count")  # Label for the y-axis

# Rotate x-axis labels if they are long or overlapping
plt.xticks(rotation=45)  # Rotate labels by 45 degrees for readability

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()  # Render the chart on the screen
```

## 6. Visualizing Numerical Data with Histograms

Histograms help you understand the **distribution** of a numerical variable (e.g., scores, ages, income):

- Is the distribution **symmetric** or **skewed**?  
- Are there **multiple peaks**?  
- Are there **outliers** (very large or small values)?


```
# Plot a histogram for the selected numerical column
plt.figure(figsize=(8, 4))  # Set size of the figure

# Use seaborn's histplot to plot the distribution with an optional KDE curve
sns.histplot(df[num_col], kde=True)  # kde=True adds a smooth density curve

# Add title and axis labels
plt.title(f"Histogram of {num_col}")  # Title of the histogram
plt.xlabel(num_col)  # Label for the x-axis
plt.ylabel("Frequency")  # Label for the y-axis

# Display the plot
plt.tight_layout()  # Adjust layout to avoid cutting off labels
plt.show()  # Render the histogram
```

## 7. Exploring Relationships with Scatter Plots

Scatter plots help you investigate relationships between **two numerical variables**, for example:

- Hours studied vs. exam score  
- Age vs. income  
- Price vs. quantity sold  


```
# Choose two numerical columns for the scatter plot
# Replace 'Score' and 'Hours_Studied' with real numerical column names
x_col = "Hours_Studied"   # TODO: change to a real numerical column (x-axis)
y_col = "Score"           # TODO: change to a real numerical column (y-axis)

# Create a scatter plot
plt.figure(figsize=(8, 5))  # Set the figure size

# Plot x_col on x-axis and y_col on y-axis
plt.scatter(df[x_col], df[y_col])  # Each point represents one row in the dataset

# Add a title and labels
plt.title(f"Scatter Plot of {y_col} vs {x_col}")  # Title describes the relationship
plt.xlabel(x_col)  # Label for x-axis
plt.ylabel(y_col)  # Label for y-axis

# Display the plot
plt.tight_layout()  # Adjust layout to avoid label cutoff
plt.show()  # Render the scatter plot
```

```
# Optionally: compute the correlation between the two numerical variables
correlation_value = df[[x_col, y_col]].corr().iloc[0, 1]  # Extracts the correlation coefficient

# Print the correlation result
print(f"Correlation between {x_col} and {y_col}: {correlation_value:.3f}")  # Shows strength and direction of relationship
```
